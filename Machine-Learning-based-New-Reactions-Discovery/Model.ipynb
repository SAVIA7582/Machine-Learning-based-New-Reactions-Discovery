# Preprocessing

## Input Data

```python
import numpy as np
import pandas as pd

from pipe import select, chain, where
```

```python
df = pd.read_csv('dataset.csv')  # dataset for training
```

```python
# original reaction smiles
original_smiles_list = df['Reaction_SMILES'].tolist()  

# 10 properties of each reaction, including reactant-product energy differences, mass-weighted coordinates, 
# dipole moments, electronegativity, molecular volume, surface polarity, 
# functional group electronic effects and electrostatic charge data
sup_df = df.drop(columns='Reaction_SMILES') 

# yield of reactions
label_df = df['y']
```

## Augment Data

```python
from AugmentData import to_augment_data, to_augment_dict, to_df

augmented_df = list(

    original_smiles_list | select(to_augment_data)  # generate augmented datas for each smiles
                         | select(to_augment_dict)  # combine datas of one smile into a dict
                         | chain
    
) | to_df()

augmented_smiles_list = augmented_df['Augmented_SMILES'].tolist()
```

```python
augmented_df.to_csv("augmented_smiles.csv", index=False)

print(augmented_df.shape)
```

## Encode Reactions

### Merge Original and Augmented Data

```python
smiles_list = augmented_smiles_list + original_smiles_list + predict_smiles_list
```

### Encode Reaction Smiles to Sequences 

```python
input_dim = 150
```

```python
from EncodeReaction import to_reaction, to_lim_int_seq, encode_properties
    
encoded_reactions = list(
    
    smiles_list | select(to_reaction)  # Convert smiles to reaction
                | where(lambda x: x is not None)  # drop invalid reactions
                | select(lambda x: to_lim_int_seq(x, input_dim-10) + encode_properties(x))  # Sequence consists of 2 parts
    
)
```

## VAE Decode

```python
latent_dim = 20
```

### Create Encoder and Decoder Pipe

```python
from VAE import predictor

# pridiction function
autoencoder_predict, vae_predict, generator_predict = predictor(input_dim, latent_dim)

def to_encoded_sequence(sequence):
    return sequence | autoencoder_predict() | vae_predict() 

def to_decoded_sequence(sequence):
    return sequence | generator_predict()
```

```python
encoded_sequences = [to_encoded_sequence(seq) for seq in encoded_reactions]
decoded_sequences = [to_decoded_sequence(seq) for seq in encoded_sequences[len(augmented_smiles_list):]]

decoded_sequences_df = decoded_sequences | to_df()
```

```python
print(decoded_sequences_df.shape)
```

## PCA Reduce Dimension

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaled_data = StandardScaler().fit_transform(decoded_sequences_df)  # Data standardization
reduced_data = PCA(n_components=90).fit_transform(scaled_data)  # PCA reduce dimensions
reduced_df = pd.DataFrame(reduced_data)
```

```python
print(reduced_df.shape)
```

## Merge with Supplemental Data

```python
merged_df = pd.concat([reduced_df, sup_df], axis=1)
```

```python
print(merged_df.shape)
```

## Filter Dimensions with Correlation Matrix

### Sample

```python
sampled_df = merged_df.sample(n=6, random_state=42)  # Select 6 rows from merged datas randomly
```

### Correlation Matrix

```python
threshold = 0.3
```

```python
# Calculate Pearson Correlation Coefficients
correlation_matrix = sampled_df.corr(method='pearson')

# screen out the correlation coefficients less than threshold
high_correlation = correlation_matrix[(correlation_matrix > threshold) & (correlation_matrix != 1.0)]
```

```python
with pd.ExcelWriter('cor-re.xlsx') as writer:
    correlation_matrix.to_excel(writer, sheet_name='Correlation Matrix')
    high_correlation.to_excel(writer, sheet_name='High Correlation')
```

### Data Filtering

```python
filtered_df = merged_df[high_correlation.dropna(how='all', subset = ['y']).index.tolist()]
```

```python
print(filtered_df.shape)
```

## Split and Convert to Tensor

### Convert to Ndarray

```python
X = np.array(filtered_df)
y = np.array(label_df) 
```

### Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
print(X_train.shape, X_test.shape)
```

### Convert to Tensor

```python
from torch import Tensor
from torch.utils.data import TensorDataset, DataLoader

def to_tensor(ndarray):
    return Tensor(ndarray).unsqueeze(1) 
```

```python
X_train_ts = to_tensor(X_train)
y_train_ts = to_tensor(y_train)
X_test_ts = to_tensor(X_test)

dataset_train = TensorDataset(X_train_ts, y_train_ts)
data_iter_train = DataLoader(dataset_train, batch_size=32, shuffle=True)
```

## Training

```python
epochs = 500
```

```python
from CNN import ConvNet
from torch.optim import Adam
import torch.nn as nn

dim = X.shape[1]  # dimension of X 

net = ConvNet(dim)
criterion = nn.MSELoss()
optimizer = Adam(net.parameters(), lr=0.001)
```

```python
loss_train = []

for epoch in range(epochs):
    
    net.train()
    
    running_loss = 0.0
    
    for X, y in data_iter_train:
        
        loss = criterion(net(X), y)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    running_loss += loss.item()
    
    print(epoch, running_loss / len(data_iter_train))
    
    loss_train.append(running_loss / len(data_iter_train))
```

```python
%matplotlib inline

from Plot import draw_train_loss

draw_train_loss(loss_train)
```
